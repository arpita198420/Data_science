                            Mean

```python
##mean:it is commonly refered as average is a measure of central tedency that represents
#typical value in dataset.it is caluclated by sum of all values divided by number of values

##mean of volume
print("Mean of volume is:",df['Volume'].mean())

##Mean of Avg Price
print("Mean of Average_price is:",df['Avg Price'].mean())

##Mean of Total Sales Value
print("Mean of Total Sales Value:",df['Total Sales Value'].mean())

#Mean of Siscount Rate %
print("Mean of Discount Rate % is :",df['Discount Rate (%)'].mean())

##Discount Amount
print("Mean of Discount Amount is:",df['Discount Amount'].mean())

##Net Sales Volume
print("Mean of Net sales value is:",df['Net Sales Value'].mean())
```

Standard Deviation

```python
##StdDeviation: It is denoted by Sigma for apopulation and S for samples
##It is a stastical measure of how sreadout values are around the mean
## A small standard deviation means values are clustered around the mean while large deviation indicates
## greater dispersion

# Standrad deviation of Volume
print("Standard deviation of Volume is:",df['Volume'].std())

# Standard deviation of Average Price
print("Standard deviation of Average Price:",df['Avg Price'].std())

#Standrd Deviation of Total Sales Valu
print("Standard deviation of Total sales value:",df['Total Sales Value'].std())

#Standrad Deviation of Discount Rate %
print("Standrad deviation of Discount rate % is:",df['Discount Rate (%)'].std())

#standrad Deviation of Net Sales Value
print("Standard deviation of Netsales volume is:",df['Net Sales Value'].std())

#Standard Deviation of Discount Amount
print("Standard deviation of Discount maount is:",df['Discount Amount'].std())
```

Medain

```python
##Median: Median is the value divides an ordered dtaaset into two equal halves.Meaning 50% are above it
## and 50% are below it.It serves to measure of central tedency that is less sensitive to ouliers
## or skwed data compared to mean

# Median of Volume
print("Median of Volume is:",df['Volume'].median())

#Median of Average Price
print("Median of Average price is:",df['Avg Price'].median())

##Median of Total Sales Value
print("Median of Total Sales Value is:",df['Total Sales Value'].median())

##Median of Discount Rate %
print("Median of Discount Rate% is:",df['Discount Rate (%)'].median())

#Median of Net Sales Value
print("Median of Net sales value is :",df['Net Sales Value'].median())

#Median of Discount
print("Median of Discount is:",df['Discount Amount'].median())
```

                               Mode

```python
#Mode :is the value that accour most frequently in dataset
# a data set can have one mode(unimode),two mode(Bimode),Threemode(Tri Mode)
##  more than three mode(multimode)
# It is useful for Categorical data also applies numerical data as well.

# Mode of Volume
print("Mode of Volume is:",df['Volume'].mode())

#Median of Average Price
print("Mode of Average price is:",df['Avg Price'].mode())

##Median of Total Sales Value
print("Mode of Total Sales Value is:",df['Total Sales Value'].mode())

##Median of Discount Rate %
print("Mode of Discount Rate% is:",df['Discount Rate (%)'].mode())

#Median of Net Sales Value
print("Mode of Net sales value is :",df['Net Sales Value'].mode())

#Median of Discount
print("Mode of Discount is:",df['Discount Amount'].mode())


```

Standardization of Numerical Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
num_cols=df.select_dtypes(['int64','Float64']).columns.tolist()
print(num_cols)
plt.figure(figsize=(16,12))
for i,cols in enumerate(num_cols,1):
  plt.subplot(3,2,i)
  sns.histplot(df[cols],kde=True,bins=30,color='green')
  plt.title(f"original distribution:{cols}")
plt.tight_layout()
plt.show()
```

```python
#Using Standardization Method
#Standardize using z=x-mu/sigma
sc=StandardScaler()
df_standardized=df.copy()
df_standardized[num_cols]=sc.fit_transform(df[num_cols])
df_standardized[num_cols].describe()
#plot using Standardized distribution
plt.figure(figsize=(16,12))
for i,col in enumerate(num_cols,1):
  plt.subplot(3,2,i)
  sns.histplot(df_standardized[col],kde=True,color='Blue',bins=20)
  plt.title(f"standard distribution: {col}")
plt.tight_layout()
plt.show()
```
